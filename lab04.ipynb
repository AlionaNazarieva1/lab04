```python
import numpy as np
import math
from scipy.integrate import odeint
import matplotlib.pyplot as plt
w = math.sqrt(3.3)
g = 7.7
t0 = 0
tmax = 55
dt = 0.05
def f(t):
    f = 0.0*math.sin(4*t)
    return f
def dy(s,t):
    dy1 =  s[1]
    dy2 =  -w*w*s[0] - g*s[1] - f(t)
    return [dy1, dy2]
t = np.arange(t0,tmax,dt) 
v0=[1.1,0]
s = odeint(dy,v0,t)
plt.plot(s,'r--', linewidth=2.0,label="решение")
plt.grid()
plt.show()
```

```python
x1=np.array(s)
```

```python
plt.plot(x1[:,0],x1[:,1],'r--', linewidth=2.0,label="фазовый портрет")
plt.legend()
plt.grid()
plt.show()
```
